### Объединение таблиц `pd.DataFrame`

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и попытавшись найти строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
rr = recipes[['id', 'name']].merge(reviews[['user_id', 'rating', 'recipe_id']], how='left', left_on='id', right_on='recipe_id').drop(columns = 'recipe_id')
rr
```

```python
rr.dropna(axis='index', how='any', subset=['rating'], inplace=True)
rr.iloc[191,:]
```

У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0.
```python
rr2 = recipes[['name', 'id']].merge(reviews.groupby('recipe_id').agg({'review': 'count'}), how='left', left_on='id', right_on='recipe_id').rename(columns={'review': 'review_count'})
rr2
```

5.3. Выясните, рецепты, добавленные в каком году, имеют наименьший средний рейтинг?

```python
rr3 = recipes[['id', 'submitted']].merge(reviews[['recipe_id', 'rating']], how='right', left_on='id', right_on='recipe_id').drop(columns=['recipe_id'])
rr3.groupby([pd.DatetimeIndex(rr3.submitted).year]).agg({'rating': 'mean'}).sort_values(by='rating')
```
