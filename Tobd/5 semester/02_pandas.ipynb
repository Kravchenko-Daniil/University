```python
data_452 = data_43[['Date', 'Volume']].merge(
    data_451_WAT[['Date', 'Volume']],
    how='outer',
    left_on='Date',
    right_on='Date',
    suffixes = ['_ALTR', '_WAT']
)
data_452['Difference'] = (data_452.Volume_WAT - data_452.Volume_ALTR).abs()
data_452.sort_values('Date', inplace=True)
data_452
```
