Количество и названия константных и уникальных полей

```python
maskUniq = myDataFrame.nunique() == myDataFrame.shape[0]
colUniq = myDataFrame.nunique()[maskUniq].index.values
print('Количество уникальных полей: ', maskUniq.sum(),
      f'\nУникальные поля\n{colUniq}', )

maskConst = myDataFrame.nunique() == 1
colConst = myDataFrame.nunique()[maskConst].index.values
print('Количество константных полей: ', maskConst.sum(),
      f'\nКонстантные поля\n{colConst}', )
```

Удалить из набора поля, не несущие информативной нагрузки - содержащие уникальные либо константные значения

```python
myDataFrame = myDataFrame.drop(np.concatenate((colUniq, colConst)), axis=1) 
myDataFrame
```

Заменить возраст пассажиров на год рождения

```python
import datetime
now = datetime.datetime.now().year
```

```python
myDataFrame['Age'] = now - myDataFrame['Age']
myDataFrame
```

Переименовать все названия колонок на русском языке

```python
newColNames = {
    'Survived': 'Выжившие',
    'Pclass': 'КлассПассажира',
    'Sex': 'Пол',
    'Age': 'Возраст',
    'SibSp': 'Сибсп',
    'Parch': 'Пергамент',
    'Ticket': 'Билет',
    'Fare': 'Стоимость',
    'Cabin': 'Купе',
    'Embarked': 'Погружены',
}
myDataFrame.rename(columns=newColNames, inplace=True)
myDataFrame
```

Заменить все отсутствующие данные средним значением (в случае числового поля) либо самым распространенным в категориальных полях.

```python
valuesToReplace = myDataFrame.mode().loc[0]
meamVal = myDataFrame.describe().loc['mean']
valuesToReplace[meamVal.index] = meamVal.values
valuesToReplace
```

```python
maskNAN = myDataFrame.isna().sum(axis=1) != 0
myDataFrame[maskNAN]
```

```python
myDataFrame = myDataFrame.fillna(valuesToReplace)
```
